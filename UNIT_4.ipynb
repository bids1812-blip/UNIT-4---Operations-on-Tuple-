```python
#Add tuple tp a tuple
thistuple = ("apple", "banana", "cherry")
y = ("orange",)   #If you don't use comma after this, it would be an ERROR
thistuple+=y
print(thistuple)
```

```python
#Address verification after appending new item in the tuple
thistuple = ("apple", "banana", "cherry")
print("Before:", id(thistuple))

y = ("Orange",)
thistuple+=y

print("After:", id(thistuple))
```
